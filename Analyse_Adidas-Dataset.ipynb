## Data Cleansing

```python
DF.head(5)
```

```python
## Delete the duplicates
DF = DF.drop_duplicates()
```

```python
cat_col
```

Correct inconsistent data 

```python
DF['Population'] = DF['Population'].astype('category')
DF['Retailer ID'] = DF['Retailer ID'].astype('category')
DF['Units Sold'] = DF['Units Sold'].str.replace(',', '') ## Delete the , from the nums
DF['Units Sold'] = pd.to_numeric(DF['Units Sold'])
DF.info()
```

```python
DF['Total Sales'] = DF['Price per Unit'] * DF['Units Sold']
```

## Feature Engineering

```python
DF.head(10)
```

```python
# convert date to monthes
DF['Invoice Date'] = pd.to_datetime(DF['Invoice Date'], dayfirst = True) # convert to date time
# DF['Invoice Date (By_Month)'] = DF['Invoice Date'].dt.month # Extract month as a number(1–12)
DF['Invoice Date (By_Year)'] = DF['Invoice Date'].dt.year
DF['Invoice Date (By_Month)'] = DF['Invoice Date'].dt.strftime('%B') # Extract month as names
# convert date to seasons
def Seasons(date):
    if date in ['December', 'January', 'February']:
        return 'Winter'
    elif date in ['March', 'April', 'May']:
        return 'Spring'
    elif date in ['June', 'July', 'August']:
        return 'Summer'
    else:
        return 'Autumn'
DF['Invoice Date (By_Season)'] = DF['Invoice Date (By_Month)'].apply(Seasons)
DF.drop('Invoice Date', axis = 1, inplace = True)
DF
```

```python
DF['Invoice Date (By_Year)'] = DF['Invoice Date (By_Year)'].astype('category')
DF.info()
```
